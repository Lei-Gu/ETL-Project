### Extract CSVs into DataFrames

```python
inspections = "Resources/food-inspections.csv"
```

```python
licenses = "Resources/business-licenses.csv"
```

```python
owners = "Resources/business-owners.csv"
```

### Transform inspections DataFrame

```python
inspections_df = pd.read_csv(inspections)
inspections_df.head()
```

```python
new_inspections_df = inspections_df[['Inspection ID', 'License #', 'DBA Name','AKA Name','Risk','Inspection Date','Inspection Type','Results','Violations']].copy()
new_inspections_df.head()
```

## Transform licenses DataFrame

```python
licenses_df = pd.read_csv(licenses)
licenses_df.head()
```

```python
new_licenses = licenses_df[
    ['ID','LICENSE ID','ACCOUNT NUMBER','LEGAL NAME','DOING BUSINESS AS NAME','ADDRESS','CITY','STATE','ZIP CODE','LICENSE CODE','LICENSE DESCRIPTION','LICENSE TERM START DATE','LICENSE TERM EXPIRATION DATE','LICENSE STATUS','LATITUDE','LONGITUDE']].copy()
new_licenses.head()
```

### Transform owners DataFrame

```python
owners_df = pd.read_csv(owners)
owners_df.head()
```

```python
new_owners_df = owners_df[
    ['Account Number','Legal Name','Owner First Name','Owner Last Name','Title']].copy()
new_owners_df.head()
```

## Transformation


```python
inspections_rename = new_inspections_df.rename(
    columns={'Inspection ID':'id', 'License #':'license_id', 'DBA Name':'dba_name','AKA Name':'aka_name','Risk':'risk','Inspection Date':'inspection_date','Inspection Type':'inspection_type','Results':'results','Violations':'violations'})
```

```python
# inspections_rename.set_index("id", inplace =True)
inspections_rename.head()
```

```python
licenses_rename=new_licenses.rename(columns={
    'ID': 'id',
    'LICENSE ID':'license_id',
    'ACCOUNT NUMBER':'account_number',
    'LEGAL NAME':'legal_name',
    'DOING BUSINESS AS NAME':'business_name',
    'ADDRESS':'address',
    'CITY':'city',
    'STATE':'state',
    'ZIP CODE':'zip_code',
    'LICENSE CODE':'license_code',
    'LICENSE DESCRIPTION':'license_description',
    'LICENSE TERM START DATE':'license_start_date',
    'LICENSE TERM EXPIRATION DATE':'lincese_end_date',
    'LICENSE STATUS':'license_status',
    'LATITUDE':'latitude',
    'LONGITUDE':'longitude'
})
```

```python
licenses_rename.set_index("id", inplace =False)
licenses_rename.head()
```

```python
owners_rename=new_owners_df.rename(
    columns={
        'Account Number': 'account_number',
        'Legal Name': 'legal_name',
        'Owner First Name': 'owner_first_name',
        'Owner Last Name': 'owner_last_name',
        'Title': 'title'}
)
owners_rename.head()
```

# Cleaning data -Licenses


```python
# check missing data
licenses_rename.count()
```

```python
# clean up the duplicate license ids if exist
licenses_cleaned = licenses_rename.dropna(subset=['license_id'])
licenses_cleaned.count()
```

```python
# seclect city as Chicago
licenses_chicago =licenses_cleaned.loc[licenses_cleaned['city'] == 'CHICAGO']

licenses_chicago.count()
```

```python
# Seperate the data into two dataframes

chicago_business_info = licenses_chicago[['account_number','legal_name','business_name','address','city','state','zip_code','latitude','longitude']].copy()
chicago_business_info.head()
```

```python
chicago_business_info.drop_duplicates(subset="account_number",inplace= True)
```

```python
chicago_business_licenses = licenses_chicago[['license_id','account_number','license_code','license_description','license_start_date','lincese_end_date','license_status']].copy()
chicago_business_licenses.head()
```
